```python
def extract_features_and_target(dataframe, target_column):
    X = dataframe.drop(columns=[target_column])
    y = dataframe[target_column]
    return X, y
```

```python
def calculate_num_features(dataframe, percent):
    num_features = (len(dataframe.columns) - 1) * percent / 100
    return math.ceil(num_features)
```

```python
def compute_relevance(X, y):
    mutual_info = mutual_info_classif(X, y)
    mutual_info_dict = dict(zip(X.columns, mutual_info))
    mutual_info_df = pd.DataFrame(list(mutual_info_dict.items()), columns=['Feature', 'Mutual Information'])
    mutual_info_df.to_csv('mutual_info_csv.csv', index=False)
    return mutual_info_dict
```

```python
def calculate_redundancy(X, selected_features, feature, mutual_info_dict):
    redundancy_value = 0
    for selected in selected_features:
        mi_fi_fs = mutual_info_regression(X[[feature]], X[selected])[0]
        mi_fi_c = mutual_info_dict[feature]  # I(C; fi)
        redundancy_value += mi_fi_fs / mi_fi_c if mi_fi_c != 0 else 0
    return redundancy_value / len(selected_features)
```

```python
def select_next_feature(X, selected_features, relevance, remaining_features, results):
    best_feature = None
    best_score = -np.inf

    for feature in remaining_features:
        redundancy = calculate_redundancy(X, selected_features, feature, relevance)
        score = relevance[feature] - redundancy
        if score > best_score:
            best_score = score
            best_feature = feature
    results.append({
                'Feature': best_feature,
                'Mutual Information': relevance[best_feature],
                'Redundancy': relevance[best_feature]-best_score,
                'GMI': best_score
            })
    return best_feature
```

```python
def fmifs(dataframe, target_column, output_csv='fmifs_results.csv'):
    X, y = extract_features_and_target(dataframe, target_column)
    relevance = compute_relevance(X, y)
    results = []
    first_feature=max(relevance, key=relevance.get)
    selected_features = [first_feature]
    results.append({
        'Feature': first_feature,
        'Mutual Information': relevance[first_feature],
        'Redundancy': 0,
        'GMI': relevance[first_feature]
    })
    remaining_features = list(set(X.columns) - set(selected_features))
    while len(remaining_features)!=0:
        best_feature = select_next_feature(X, selected_features, relevance, remaining_features, results)
        selected_features.append(best_feature)
        remaining_features = list(set(X.columns) - set(selected_features))

    results_sorted = sorted(results, key=lambda x: x['GMI'], reverse=True)\

    results_df = pd.DataFrame(results_sorted)
    results_df.to_csv(output_csv, index=False)

    sorted_features = [result['Feature'] for result in results_sorted]
    return sorted_features
```

```python
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import VotingClassifier
```

```python
def svm(X_train, y_train, X_test, y_test):
    svm_clf = LinearSVC(random_state=0, tol=1e-5)
    svm_clf.fit(X_train, y_train)
    y_pred=svm_clf.predict(X_test)
    return accuracy_score( y_test, y_pred )
```

```python
def naive_bayes(X_train, y_train, X_test, y_test):
    gnb_classifier = GaussianNB()
    gnb_classifier.fit(X_train, y_train)
    y_pred_gnb = gnb_classifier.predict(X_test)
    return accuracy_score(y_test, y_pred_gnb)
```

```python
def decision_tree(X_train, y_train, X_test, y_test):
    dt_classifier = DecisionTreeClassifier()
    dt_classifier.fit(X_train, y_train)
    y_pred_dt = dt_classifier.predict(X_test)
    return accuracy_score(y_test, y_pred_dt)
```

```python
def xgboost(X_train, y_train, X_test, y_test):
    xgb_classifier = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
    xgb_classifier.fit(X_train, y_train)
    y_pred_xgb = xgb_classifier.predict(X_test)
    return accuracy_score(y_test, y_pred_xgb)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import time
def extraTree(X_train, y_train, X_test, y_test):
    start = time.time()
    model = ExtraTreesClassifier(random_state=0,n_jobs=-1).fit(X_train,y_train)
    end_train = time.time()
    y_predictions = model.predict(X_test) # These are the predictions from the test data.
    end_predict = time.time()
    return accuracy_score(y_test, y_predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier
def randomForest(X_train, y_train, X_test, y_test):
    start = time.time()
    model = RandomForestClassifier(n_estimators = 100,n_jobs=-1,random_state=0,bootstrap=True,).fit(X_train,y_train)
    end_train = time.time()
    y_predictions = model.predict(X_test) # These are the predictions from the test data.
    end_predict = time.time()
    return accuracy_score(y_test, y_predictions)
```

```python
from sklearn.ensemble import VotingClassifier
def ensemble(X_train, y_train, X_test, y_test):
    dt_classifier = DecisionTreeClassifier()
    xgb_classifier = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
    extra_tree = ExtraTreesClassifier(random_state=0,n_jobs=-1)
    random_forest = RandomForestClassifier(n_estimators = 100,n_jobs=-1,random_state=0,bootstrap=True)
    voting_classifier = VotingClassifier(estimators=[
        ('dt', dt_classifier),
        ('xgb', xgb_classifier),
        ('et', extra_tree),
        ('rf', random_forest),
    ], voting='hard')
    voting_classifier.fit(X_train, y_train)
    y_pred_ensemble = voting_classifier.predict(X_test)
    accuracy_ensemble = accuracy_score(y_test, y_pred_ensemble)
    return accuracy_ensemble
```

```python
def accuracy(train_data, test_data, top_features, target, func):
    X_train = train_data[top_features]
    y_train = train_data[target]
    X_test = test_data[top_features]
    y_test = test_data[target]
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    return func(X_train, y_train, X_test, y_test)
```
